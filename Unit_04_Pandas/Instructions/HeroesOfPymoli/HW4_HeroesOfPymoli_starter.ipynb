```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

# Display first 10 rows
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
purchase_data_subset = purchase_data.loc[:,["Gender","SN","Age"]]
purchase_unique_players_data = purchase_data_subset.drop_duplicates()
purchase_unique_players= purchase_data_subset.drop_duplicates().count()[0]
purchase_unique_players
summary_table = pd.DataFrame({"Total Players": [purchase_unique_players]})
summary_table
#purchase_unique_players_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Mean Price
mean_price=purchase_data['Price'].mean()

# Total Number of Purchases
total_purchases=purchase_data['Price'].count()

# Total Revenue
total_revenue=purchase_data['Price'].sum()

# Total Number of unique items
unique_items=len(purchase_data['Item ID'].unique())

# Summary Data Frame
summary_table = pd.DataFrame({"Mean Price": [mean_price],
                              "Number of Purchases": [total_purchases],
                              "Total Revenue": [total_revenue],
                              "Number of Unique Items": [unique_items]})

#summary_table.style.format({'Mean Price':'${:,.2f}','Total Revenue':'${:,.2f}'})

summary_table["Mean Price"] = summary_table["Mean Price"].map("${:,.2f}".format)
summary_table["Total Revenue"] = summary_table["Total Revenue"].map("${:,.2f}".format)
summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Group by Gender
gender = purchase_data.groupby("Gender")

# Count unique number of players
total_count_gender = gender.nunique()["SN"]

# Total count by each gender/divivde by total players 
percentage_gender = total_count_gender / len(purchase_data["SN"].value_counts()) * 100

# Create data frame with obtained values
summary_table = pd.DataFrame({"Percentage of Players": percentage_gender, "Total Count": total_count_gender})

# Format the data frame with no index name in the corner
summary_table.index.name = None
                         
# Format the values sorted by total count in descending order, and two decimal places for the percentage
summary_table.sort_values(["Total Count"], ascending = False).style.format({"Percentage of Players":"{:.2f}%"})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Total Number of Purchases per gender
total_purchases=purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")

# Average Price paid per gender
mean_purchases=purchase_data.groupby(["Gender"]).mean()["Price"].rename("Mean Purchase Value")

# Number of Purchases per gender
number_purchases=purchase_data.groupby(["Gender"]).count()["Price"].rename("Purchase Count")

# Average Total Purchase per Person
mean_purchases_person = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")/total_count_gender

# Summary Data Frame
summary_table = pd.DataFrame({"Purchase Count": number_purchases, "Average Purchase Price": mean_purchases, "Total Purchase Value": total_purchases, "Avg Total Purchase per Person": mean_purchases_person})

# Format Summary Data Frame
summary_table["Average Purchase Price"] = summary_table["Average Purchase Price"].map("${:,.2f}".format)
summary_table["Total Purchase Value"] = summary_table["Total Purchase Value"].map("${:,.2f}".format)
summary_table["Purchase Count"] = summary_table["Purchase Count"].map("{:,}".format)
summary_table["Avg Total Purchase per Person"] = summary_table["Avg Total Purchase per Person"].map("${:,.2f}".format)
summary_table = summary_table.loc[:, ["Purchase Count", "Average Purchase Price", "Total Purchase Value","Avg Total Purchase per Person"]]

# Display the Summary Table
summary_table
```

* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99,34.99,39.99,1000000.99]
bin_name = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","+40"]
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins=age_bins, labels=bin_name)



age_bins_total_repeat = purchase_data["Age Ranges"].value_counts()
age_bins_total

# Total Purchases price per age group
age_purchase_total = purchase_data.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Value")

# Average Purchase Price per age group
age_purchase_mean = purchase_data.groupby(["Age Ranges"]).mean()["Price"].rename("Total Purchase Value")





# Average Total Purchase per Person
age_bins_total = purchase_unique_players_data["Age Ranges"].value_counts()
age_purchase_mean_per_person = age_purchase_total/age_bins_total

# Summary Data Frame
summary_table = pd.DataFrame({"Purchase Count": age_bins_total_repeat, "Average Purchase Price": age_purchase_mean,
                             "Total Purchase Value": age_purchase_total,"Avg Total Purchase per Person": age_purchase_mean_per_person})

# Format Summary Data Frame

summary_table["Average Purchase Price"] = summary_table["Average Purchase Price"].map("${:,.2f}".format)
summary_table["Total Purchase Value"] = summary_table["Total Purchase Value"].map("${:,.2f}".format)
summary_table["Avg Total Purchase per Person"] = summary_table["Avg Total Purchase per Person"].map("${:,.2f}".format)
summary_table = summary_table.loc[:, ["Purchase Count", "Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]]
                             

# Display the Summary Table
summary_table = summary_table.sort_index()
summary_table
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group by SN
sn_group = purchase_data.groupby("SN")

# Total Purchases based on SN
sn_purchases = sn_group["Purchase ID"].count()

# Calculate the average purchase by name
avg_purchase = sn_group["Price"].mean()

# Calculate purchase total
purchase_total = sn_group["Price"].sum()

# Create data frame with obtained values
top_spenders = pd.DataFrame({"Purchase Count": sn_purchases,
                             "Average Purchase Price": avg_purchase,
                             "Total Purchase Value":purchase_total})

# Sort in descending order to obtain top 5 spender names
formatted_spenders = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()

# Format with currency style
formatted_spenders.style.format({"Average Purchase Total":"${:,.2f}",
                                 "Average Purchase Price":"${:,.2f}",
                                 "Total Purchase Value":"${:,.2f}"})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# New data frame
popular_items = purchase_data[["Item ID", "Item Name", "Price"]]

# Group by item id and item name
popular_items = popular_items.groupby(["Item ID","Item Name"])

# Number of times an item has been purchased
popular_items_count = popular_items["Price"].count()

# Purchase value per item
popular_items_purchase = (popular_items["Price"].sum())

# Individual item price
individual_price = popular_items_purchase/popular_items_count

# Summary Table
summary_table = pd.DataFrame({"Purchase Count": popular_items_count,
                                   "Item Price": individual_price ,
                                   "Total Purchase Value":popular_items_purchase})

# Summary Table
summary_table1 = summary_table.sort_values(["Purchase Count"], ascending=False).head(5)
#summary_table2 = summary_table.sort_values(["Total Purchase Value"], ascending=False).head(5)
# Format with currency style
summary_table1.style.format({"Item Price":"${:,.2f}","Total Purchase Value":"${:,.2f}"})
                                
```
